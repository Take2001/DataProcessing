```python
import pandas as pd
import pyexcel
import xlrd
import xlsxwriter
import os

# best function ever

def run():
    
    # some vars
    cwd = os.getcwd()
    c = 1
    concat_list = []
    
    #loop over folders in folder
    for root, dirs, files in os.walk('.'):
        
        #loop over folders in folders 
        for folder in dirs:
            print(type(folder), folder)
            
            # dont go in this one
            if folder == ".ipynb_checkpoints":
                continue
            
            # if somehow it does happen go back one folder
            if os.getcwd() == ".ipynb_checkpoints":
                os.chrdir('..')
                continue
            
            # change directory to country
            os.chdir(folder)
            
            # loop over files (tuple)
            for files in os.walk('.'):
                
                print('changed dir to {} os walk in country file'.format(folder))
                
                # store the third item in the tuple as kazan (random chosen name)
                kazan = files[2]
                
                # kazan is a list with two items
                # an xlsx and csv
                for file in kazan:
                    
                    print("selecting correct file type -> .CSV")
                    
                    # first .csv found append
                    if file.endswith(".csv") and c == 1:
                        
                        print('append CSV with first line')
                        
                        # make this loop unaccesable from this point
                        c += 1
                        csv = pd.read_csv(file)
                        
                        concat_list.append(csv)
                        
                        # go back one dir
                        os.chdir("..")
                        
                    # all other csv append without first line
                    elif file.endswith(".csv"):
                        
                        print('append file without the first line')
                            
                        csv = pd.read_csv(file, skiprows=0)
                        
                        concat_list.append(csv)
                        
                        os.chdir("..")
    
    # make sure you're back in the original file you started iterating in
    if os.getcwd() == cwd:
        
        # concat the list
        tas_pr_1991_2015_AC = pd.concat(concat_list)
        
        #return the .csv product of run()
        return tas_pr_1991_2015_AC.to_csv("tas_pr_1991_2015_AC.csv", encoding="utf-8", index=True)
```

```python
run()
```
