```python
import pandas as pd
import pyexcel
import xlrd
import xlsxwriter
import os

def run():
    
    cwd = os.getcwd()
    c = 1
    concat_list = []
    
    for root, dirs, files in os.walk('.'):
        
        for folder in dirs:

            os.chdir(folder)

            for files in os.walk('.'):

                kazan = files[2]
                
                for file in kazan:
                    
                    if file.endswith(".csv") and c == 1:

                        c += 1
                        csv = pd.read_csv(file)
                        
                        concat_list.append(csv)

                    elif file.endswith(".csv"):
                        
                        csv = pd.read_csv(file, skiprows=0)
                        
                        concat_list.append(csv)
        
                
                os.chdir("..")
                
    print(concat_list)
    
    
    if os.getcwd() == cwd:

        tas_pr_1991_2015_AC = pd.concat(concat_list)

        return tas_pr_1991_2015_AC.to_csv("tas_pr_1991_2015_AC.csv", encoding="utf-8", index=True)
```

```python
run()
```
