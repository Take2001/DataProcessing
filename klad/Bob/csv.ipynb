```python
def select_plot(input_data, country_of_residence, origin, beginyear, endyear):
    year1 = input_data
    year1 = year1.loc[year1['Country of residence'] == country_of_residence]
    year1 = year1.loc[year1['Origin'] == origin]
    year1 = year1.loc[input_data['Year'] == beginyear]
    output_data = year1
    for i in range(beginyear + 1, endyear + 1):
        print(i)
        yeari = input_data
        yeari = yeari.loc[yeari['Country of residence'] == country_of_residence]
        yeari = yeari.loc[yeari['Origin'] == origin]
        yeari = yeari.loc[input_data['Year'] == i]
        output_data = pd.concat([output_data, yeari])   
    return output_data
    
select_plot(df, 'Belgium', 'Zimbabwe', 2000, 2012)
```
