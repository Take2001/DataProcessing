```python
# kies land
me1 = fIn.loc[fIn['adm0_name'] == 'Afghanistan']
# kies markten
me1_market = me1.loc[me1['mkt_name'] == 'Fayzabad']
me2_market = me1.loc[me1['mkt_name'] == 'Hirat']
me3_market = me1.loc[me1['mkt_name'] == 'Mazar'] 
# kies producten
me1_product = me1_market.loc[me1_market['cm_name'] == 'Bread']
me2_product = me2_market.loc[me2_market['cm_name'] == 'Bread']
me3_product = me3_market.loc[me3_market['cm_name'] == 'Bread']

# plotspul
x_values = me1_product['Year']
y_values = me1_product['mp_price']
y_values2 = me2_product['mp_price']
y_values3 = me3_product['mp_price']
plt.plot(x_values, y_values)
plt.plot(x_values, y_values2)
plt.plot(x_values, y_values3)
```

```python
x_values = me1_product['Year']
y_values = me1_product['mp_price']
y_values2 = me2_product['mp_price']
plt.bar(x_values, y_values, width=0.03, label = "Hirat")
plt.bar(x_values, y_values, width=0.02,label="Fayzabad")
plt.bar(x_values, y_values3, width=0.02, label = 'Mazar')
plt.xlabel("Jaren")
plt.ylabel("Prijs in AFN")
plt.title("Broodprijs in Afghanistan")
plt.legend()
```
