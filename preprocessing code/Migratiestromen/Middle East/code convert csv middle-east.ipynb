```python
import pandas as pd
import numpy as np

raw_data = pd.read_csv('unhcr_popstats_export_asylum_seekers_monthly_all_data.csv')
df = pd.DataFrame(raw_data)

df['Origin'] = df['Origin'].replace({"Iran (Islamic Rep. of)":"Iran"})
df['Origin'] = df['Origin'].replace({"Palestinian":"State of Palestine"})


af1 = df.loc[df['Origin'] == 'Syrian Arab Republic']
af2 = df.loc[df['Origin'] == 'Afghanistan']
af3 = df.loc[df['Origin'] == 'Iraq']
af4 = df.loc[df['Origin'] == 'Lebanon']
af5 = df.loc[df['Origin'] == 'Palestinian']
af6 = df.loc[df['Origin'] == 'Pakistan']
af7 = df.loc[df['Origin'] == 'Iran']

predata_middle_east = [af1, af2, af3, af4, af5, af6, af7]

data_middle_east = pd.concat(predata_middle_east)
data_middle_east
```

```python
# vervang steretjes door nullen en maak integers van strings
data_middle_east = data_middle_east.replace(['*'], 0)
new_values = [int(row) for row in data_middle_east['Value']]

print(new_values)
```

```python
def month_converter(month):
    months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 
              'December']
    return (months.index(month))/12
```

```python
# remove old string values
data_middle_east = data_middle_east[['Country of residence', 'Origin', 'Year','Month']]

#add integer values
data_middle_east['Values'] = new_values

data_middle_east

months = []
for row in data_middle_east['Month']:
    months.append(str(row))

#print(months)
YearMonth = []

count = 0
for year in data_middle_east['Year']:
    year = int(year) + month_converter(months[count])
    YearMonth.append(year)
    count += 1

data_middle_east['YearMonth'] = YearMonth
```
